### Com os resultados em mãos, vamos responder as seguintes perguntas

#### 1. Influência do gênero

```python
gender_counts = base["Gender"].value_counts()


plt.bar(gender_counts.index, gender_counts.values, color='green')

plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')
plt.grid(linestyle='-', linewidth=0.5)
plt.show()
```

#### 2. Influência da escolaridade

```python

education_counter = base["Education"].value_counts()

plt.figure(figsize=(10, 6))

plt.bar(education_counter.index, education_counter.values, color='green')

plt.xticks(rotation=45, ha='right')

plt.xlabel('Education')
plt.ylabel('Count')
plt.title('Education grade')
plt.grid(linestyle='-', linewidth=0.5)
plt.tight_layout()
plt.show()
```

#### 3. País com mais pessoas com alto QI (Avaliado pela média de QI em cada país)

```python
compacted = base.groupby("Country")["IQ"].mean()

plt.figure(figsize=(10, 6))

plt.bar(compacted.index, compacted.values, color='green')

plt.title("Country with more IQ")
plt.ylabel("Count")
plt.grid(linestyle='-', linewidth=0.5)
plt.tight_layout()
plt.show()
```

#### 4. Quantidade de pessoas por matéria (Avaliado pela média de QI)

```python
compacted_by_field = base.groupby("Field of Expertise")["IQ"].mean()

plt.figure(figsize=(10, 6))

plt.bar(compacted_by_field.index, compacted_by_field.values, color='green')

plt.title("Field of Expertise vs Mean IQ")
plt.ylabel("Mean IQ")
plt.grid(linestyle='-', linewidth=0.5)
plt.tight_layout()
plt.show()
```
