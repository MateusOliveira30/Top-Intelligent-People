## DSA

### 1. Influence of gender

```python
gender_counts = base["Gender"].value_counts()

plt.bar(gender_counts.index, gender_counts.values, color='green')

plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')
plt.grid(linestyle='-', linewidth=0.5)
plt.show()
```

### 2. Influence of education

```python
education_counter = base["Education"].value_counts()

plt.figure(figsize=(10, 6))

plt.bar(education_counter.index, education_counter.values, color='green')

plt.xticks(rotation=45, ha='right')

plt.xlabel('Education')
plt.ylabel('Count')
plt.title('Education grade')
plt.grid(linestyle='-', linewidth=0.5)
plt.tight_layout()
plt.show()
```

### 3. Country with the most people with high IQ (Evaluated by the average IQ in each country)

```python
compacted = base.groupby("Country")["IQ"].mean()



plt.figure(figsize=(10, 6))

plt.bar(compacted.index, compacted.values, color='green')

plt.title("Country with high IQ")
plt.ylabel("Count")
plt.grid(linestyle='-', linewidth=0.5)
plt.tight_layout()
plt.show()
```
